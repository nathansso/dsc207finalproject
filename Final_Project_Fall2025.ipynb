````python
# Conduct Hypothesis Test on Weekend proportion to determine if likelihood of session occuring on weekend is significantly different from 2/7 (28.57%) at 95% confidence level
from statsmodels.stats.proportion import proportions_ztest
# Number of sessions that occurred on the weekend
weekend_sessions = ecom_df['Weekend'].sum()
# Total number of sessions
total_sessions = len(ecom_df)
# Proportion of weekend sessions
weekend_proportion = weekend_sessions / total_sessions
# Perform one-sample z-test against the null hypothesis that the true proportion is 2/7
z_stat, p_value = proportions_ztest(count=weekend_sessions, nobs=total_sessions, value=2/7, alternative='two-sided')
print(f"Weekend Proportion: {weekend_proportion:.4f}")
print(f"Z-statistic: {z_stat:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: The proportion of weekend sessions is significantly different from 2/7.")
else:
    print("Fail to reject the null hypothesis: The proportion of weekend sessions is not significantly different from 2/7.")
````

## Outlier

### Outlier Detection

```

Firstly, I want to calculate the IQR's of each column and identify outliers outside of these bounds:
    [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR]

```

First I want to conduct hypothesis testing on features individually to determine which features have a statistically significant relationship with the target variable (Revenue). 
It seems safe to assume that the observations are drawn independently.
However, the distribution of features is not normal or log-normal.
Will run Mann-Whitney U test for numeric features and Chi-Squared test for categorical features to identify significant relationships with Revenue.

The Hypothesis Test will be formatted like so:
    Null Hypothesis: There is no difference in values of Feature conditoned on Revenue and Feature conditioned on No Revenue
    Alternative Hypothesis: There is a difference in values of Feature conditoned on Revenue and Feature conditioned on No Revenue

```

````python
#First I want to conduct hypothesis testing on features individually to determine which features have a statistically significant relationship with the target variable (Revenue). 
#It seems safe to assume that the observations are drawn independently.
#However, the distribution of features is not normal or log-normal.
# Will run Mann-Whitney U test for numeric features and Chi-Squared test for categorical features to identify significant relationships with Revenue.

p_rows = []
#Hypothesis Testing for Numeric Features using Mann-Whitney U Test
from scipy.stats import mannwhitneyu

print("=== Numeric Features: Mann-Whitney U Test ===")
print(f"{'Feature':<35} {'U-stat':>12} {'p-value':>12} {'Significant?':>14}")
print("-" * 75)

rev_true = ecom_df[ecom_df['Revenue'] == True]
rev_false = ecom_df[ecom_df['Revenue'] == False]

for col in f_num_cols:
    stat, p = mannwhitneyu(rev_true[col], rev_false[col], alternative='two-sided')
    sig = "Yes" if p < 0.05 else "No"
    print(f"{col:<35} {stat:>12.1f} {p:>12.4e} {sig:>14}")
    p_rows.append({'Feature': col, 'Test': 'Mann-Whitney U', 'Statistic': stat, 'p-value': p})

#Hypothesis Testing for Categorical Features using Chi-Squared Test
from scipy.stats import mannwhitneyu, chi2_contingency
print("\n=== Categorical Features: Chi-Squared Test ===")
print(f"{'Feature':<35} {'Chi2':>12} {'p-value':>12} {'Significant?':>14}")
print("-" * 75)

cat_test_cols = [c for c in f_cat_cols if c != 'Revenue']
for col in cat_test_cols:
    contingency = pd.crosstab(ecom_df[col], ecom_df['Revenue'])
    chi2, p, dof, expected = chi2_contingency(contingency)
    sig = "Yes" if p < 0.05 else "No"
    print(f"{col:<35} {chi2:>12.2f} {p:>12.4e} {sig:>14}")
    p_rows.append({'Feature': col, 'Test': 'Chi-Squared', 'Statistic': chi2, 'p-value': p})


p_vals = pd.DataFrame(p_rows, columns=['Feature', 'Test', 'Statistic', 'p-value'])
````

````python
p_vals
alpha = 0.05
p_vals['Significant'] = p_vals['p-value'] < alpha
p_vals.sort_values('p-value', inplace=True)
p_vals
````
