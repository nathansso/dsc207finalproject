````python
# Conduct Hypothesis Test on Weekend proportion to determine if likelihood of session occuring on weekend is significantly different from 2/7 (28.57%) at 95% confidence level
from statsmodels.stats.proportion import proportions_ztest
# Number of sessions that occurred on the weekend
weekend_sessions = ecom_df['Weekend'].sum()
# Total number of sessions
total_sessions = len(ecom_df)
# Proportion of weekend sessions
weekend_proportion = weekend_sessions / total_sessions
# Perform one-sample z-test against the null hypothesis that the true proportion is 2/7
z_stat, p_value = proportions_ztest(count=weekend_sessions, nobs=total_sessions, value=2/7, alternative='two-sided')
print(f"Weekend Proportion: {weekend_proportion:.4f}")
print(f"Z-statistic: {z_stat:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: The proportion of weekend sessions is significantly different from 2/7.")
else:
    print("Fail to reject the null hypothesis: The proportion of weekend sessions is not significantly different from 2/7.")
````

### Outlier Detection

```

Firstly, I want to calculate the IQR's of each column and identify outliers outside of these bounds:
    [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR]

```

First I want to conduct hypothesis testing on features individually to determine which features have a statistically significant relationship with the target variable (Revenue). 
It seems safe to assume that the observations are drawn independently.
However, the distribution of features is not normal or log-normal.
Will run Mann-Whitney U test for numeric features and Chi-Squared test for categorical features to identify significant relationships with Revenue.

The Hypothesis Test will be formatted like so:
    Null Hypothesis: There is no difference in values of Feature conditoned on Revenue and Feature conditioned on No Revenue
    Alternative Hypothesis: There is a difference in values of Feature conditoned on Revenue and Feature conditioned on No Revenue

```

````python
#First I want to conduct hypothesis testing on features individually to determine which features have a statistically significant relationship with the target variable (Revenue). 
#It seems safe to assume that the observations are drawn independently.
#However, the distribution of features is not normal or log-normal.
# Will run Mann-Whitney U test for numeric features and Chi-Squared test for categorical features to identify significant relationships with Revenue.

p_rows = []
#Hypothesis Testing for Numeric Features using Mann-Whitney U Test
from scipy.stats import mannwhitneyu

print("=== Numeric Features: Mann-Whitney U Test ===")
print(f"{'Feature':<35} {'U-stat':>12} {'p-value':>12} {'Significant?':>14}")
print("-" * 75)

rev_true = ecom_df[ecom_df['Revenue'] == True]
rev_false = ecom_df[ecom_df['Revenue'] == False]

for col in f_num_cols:
    stat, p = mannwhitneyu(rev_true[col], rev_false[col], alternative='two-sided')
    sig = "Yes" if p < 0.05 else "No"
    print(f"{col:<35} {stat:>12.1f} {p:>12.4e} {sig:>14}")
    p_rows.append({'Feature': col, 'Test': 'Mann-Whitney U', 'Statistic': stat, 'p-value': p})

#Hypothesis Testing for Categorical Features using Chi-Squared Test
from scipy.stats import mannwhitneyu, chi2_contingency
print("\n=== Categorical Features: Chi-Squared Test ===")
print(f"{'Feature':<35} {'Chi2':>12} {'p-value':>12} {'Significant?':>14}")
print("-" * 75)

cat_test_cols = [c for c in f_cat_cols if c != 'Revenue']
for col in cat_test_cols:
    contingency = pd.crosstab(ecom_df[col], ecom_df['Revenue'])
    chi2, p, dof, expected = chi2_contingency(contingency)
    sig = "Yes" if p < 0.05 else "No"
    print(f"{col:<35} {chi2:>12.2f} {p:>12.4e} {sig:>14}")
    p_rows.append({'Feature': col, 'Test': 'Chi-Squared', 'Statistic': chi2, 'p-value': p})


p_vals = pd.DataFrame(p_rows, columns=['Feature', 'Test', 'Statistic', 'p-value'])
````

````python
p_vals
alpha = 0.05
p_vals['Significant'] = p_vals['p-value'] < alpha
p_vals.sort_values('p-value', inplace=True)
p_vals
````

I want to tune the hyperparameters (Max Depth, Minimum Leaf Samples, and Number of Estimators). 

Because in Random Forests these terms weakly interact, I will find a tight range of optimal values for each individual hyperparameter utilizing out of bag scores.

Next, I will utilize a Randomized Search Cross Validation over the listed hyperparameters with their tight ranges and these additonal features:
    - max_features
    - min_samples_split

Because my dataset is imbalanced between Revenue and Non-Revenue counts I want to use ROC-AUC scoring in my Randomized Search.

````python
# Optimized Random Forest
# Initially sweep depths using out of bag score to save time, finding optimal depth.

depths = [5, 8, 10, 12, 15, 18, 20, 25, 40, None]
oob_scores = []

for d in depths:
    rf = RandomForestClassifier(
        n_estimators=300,
        max_depth=d,
        min_samples_leaf=5,
        max_features='sqrt',
        class_weight='balanced',
        oob_score=True,
        bootstrap=True,
        random_state=42,
        n_jobs=-1
    )

    rf.fit(X_train, y_train)
    oob_scores.append(rf.oob_score_)

results = pd.DataFrame({
    "max_depth": depths,
    "oob_score": oob_scores
})

print(results)
print("Best depth:", results.loc[results.oob_score.idxmax(), "max_depth"])

plt.plot([str(d) for d in depths], oob_scores, marker='o')
plt.xlabel("max_depth")
plt.ylabel("OOB Score")
plt.title("Random Forest Depth Tuning")
plt.show()
````

```
Based on these Out Of Bag Scores, It seems that prior to a depth of 8 the model may be too simple and suffers. However, beyond the max_depth of 25 improvement in OOBS plateus. Therefore I will likely restrict the range of Max_Depth to 18 to 25, where the improvement seems to mostly level off.

```

````python
# Optimized Random Forest
# Sweep Minimum Samples Leaf using out of bag score to save time, finding optimal min_samples_leaf.

msl = [1, 2, 3, 4, 5, 10, 15, 20, 30]
oob_scores = []

for m in msl:
    rf = RandomForestClassifier(
        n_estimators=300,
        max_depth=25,
        min_samples_leaf=m,
        max_features='sqrt',
        class_weight='balanced',
        oob_score=True,
        bootstrap=True,
        random_state=42,
        n_jobs=-1
    )

    rf.fit(X_train, y_train)
    oob_scores.append(rf.oob_score_)

results = pd.DataFrame({
    "min_samples_leaf": msl,
    "oob_score": oob_scores
})

print(results)
print("Best min_samples_leaf:", results.loc[results.oob_score.idxmax(), "min_samples_leaf"])

plt.plot([str(m) for m in msl], oob_scores, marker='o')
plt.xlabel("min_samples_leaf")
plt.ylabel("OOB Score")
plt.title("Random Forest Minimum Samples Leaf Tuning")
plt.show()
````

```

Based on these results the model performs best when we create a more complex model. With the fact that increasing the minimum size, slightly decreases performance, it seems like the Random Forest isn't overfitting.

Additonally this seems to suggest that, we can have many informative splits. This makes sense with out hypothesis testing, all but one of our features was statistically significant. This means that each feature can provide meaning full information.

#### First Solution: Threshold Tuning

````python
# Threshold Tuning for Optimized Random Forest
from sklearn.metrics import precision_recall_curve, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

# Get predicted probabilities for the positive class (Revenue=True)
y_proba_rf = rf_randomized_search.predict_proba(X_test)[:, 1]

# Compute precision, recall, and thresholds
rf_precisions, rf_recalls, rf_thresholds = precision_recall_curve(y_test, y_proba_rf)

# Calculate F1 score at each threshold
rf_f1_scores = 2 * (rf_precisions[:-1] * rf_recalls[:-1]) / (rf_precisions[:-1] + rf_recalls[:-1] + 1e-10)

# Find the threshold that maximizes F1
rf_best_idx = np.argmax(rf_f1_scores)
rf_best_threshold = rf_thresholds[rf_best_idx]
rf_best_f1 = rf_f1_scores[rf_best_idx]

print(f"Best Threshold: {rf_best_threshold:.3f}")
print(f"Best F1 Score at that threshold: {rf_best_f1:.3f}")
print(f"Precision at best threshold: {rf_precisions[rf_best_idx]:.3f}")
print(f"Recall at best threshold: {rf_recalls[rf_best_idx]:.3f}")

# Plot Precision-Recall vs Threshold
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Left: Precision & Recall vs Threshold
ax[0].plot(rf_thresholds, rf_precisions[:-1], label='Precision', color='steelblue')
ax[0].plot(rf_thresholds, rf_recalls[:-1], label='Recall', color='darkorange')
ax[0].plot(rf_thresholds, rf_f1_scores, label='F1 Score', color='green', linestyle='--')
ax[0].axvline(x=rf_best_threshold, color='red', linestyle=':', label=f'Best Threshold = {rf_best_threshold:.3f}')
ax[0].set_xlabel('Threshold')
ax[0].set_ylabel('Score')
ax[0].set_title('Precision, Recall & F1 vs. Classification Threshold')
ax[0].legend()
ax[0].grid(True, alpha=0.3)

# Right: Precision-Recall Curve
ax[1].plot(rf_recalls, rf_precisions, color='steelblue')
ax[1].scatter(rf_recalls[rf_best_idx], rf_precisions[rf_best_idx], color='red', s=100, zorder=5, label=f'Best F1 ({rf_best_threshold:.3f})')
ax[1].set_xlabel('Recall')
ax[1].set_ylabel('Precision')
ax[1].set_title('Precision-Recall Curve')
ax[1].legend()
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Apply optimal threshold and show classification report
y_pred_rf_tuned = (y_proba_rf >= rf_best_threshold).astype(int)
print("\n--- Optimized RF with Tuned Threshold ---")
print(classification_report(y_test, y_pred_rf_tuned))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf_tuned))
````

### XGBoost Model

````python
# Creating an XGBoost model to compare feature importance and performance with the Random Forest model.
from xgboost import XGBClassifier

# Calculate scale_pos_weight to handle ~85/15 class imbalance (analogous to class_weight='balanced' in RF)
scale_pos = (y_train == 0).sum() / (y_train == 1).sum()
print(f"scale_pos_weight: {scale_pos:.2f}")

xgb_base = XGBClassifier(
    eval_metric='logloss',
    random_state=42
)

xgb_param_grid = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.5, 0.7, 0.8, 1.0],
    'min_child_weight': [1, 3, 5, 7],
    'gamma': [0, 0.1, 0.3, 0.5],
    'scale_pos_weight': [1, scale_pos],
}

xgb_randomized_search = RandomizedSearchCV(
    estimator=xgb_base,
    param_distributions=xgb_param_grid,
    cv=5,
    n_jobs=-1,
    scoring='f1',
    n_iter=100,
    random_state=42,
    verbose=1
)
xgb_randomized_search.fit(X_train, y_train)
````

````python
# XGBoost Results
print(f"Best Hyperparameters: {xgb_randomized_search.best_params_}")
print(f"Best CV F1 Score: {xgb_randomized_search.best_score_:.4f}\n")

y_pred_xgb = xgb_randomized_search.predict(X_test)
print("XGBoost Classification Report:")
print(classification_report(y_test, y_pred_xgb))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb))
````

````python
# Threshold Tuning for XGBoost
from sklearn.metrics import precision_recall_curve, f1_score, classification_report, confusion_matrix

# Get predicted probabilities for the positive class (Revenue=True)
y_proba_xgb = xgb_randomized_search.predict_proba(X_test)[:, 1]

# Compute precision, recall, and thresholds
xgb_precisions, xgb_recalls, xgb_thresholds = precision_recall_curve(y_test, y_proba_xgb)

# Calculate F1 score at each threshold
xgb_f1_scores = 2 * (xgb_precisions[:-1] * xgb_recalls[:-1]) / (xgb_precisions[:-1] + xgb_recalls[:-1] + 1e-10)

# Find the threshold that maximizes F1
xgb_best_idx = np.argmax(xgb_f1_scores)
xgb_best_threshold = xgb_thresholds[xgb_best_idx]
xgb_best_f1 = xgb_f1_scores[xgb_best_idx]

print(f"Best Threshold: {xgb_best_threshold:.3f}")
print(f"Best F1 Score at that threshold: {xgb_best_f1:.3f}")
print(f"Precision at best threshold: {xgb_precisions[xgb_best_idx]:.3f}")
print(f"Recall at best threshold: {xgb_recalls[xgb_best_idx]:.3f}")

# Plot Precision-Recall vs Threshold
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

ax[0].plot(xgb_thresholds, xgb_precisions[:-1], label='Precision', color='steelblue')
ax[0].plot(xgb_thresholds, xgb_recalls[:-1], label='Recall', color='darkorange')
ax[0].plot(xgb_thresholds, xgb_f1_scores, label='F1 Score', color='green', linestyle='--')
ax[0].axvline(x=xgb_best_threshold, color='red', linestyle=':', label=f'Best Threshold = {xgb_best_threshold:.3f}')
ax[0].set_xlabel('Threshold')
ax[0].set_ylabel('Score')
ax[0].set_title('XGBoost: Precision, Recall & F1 vs. Classification Threshold')
ax[0].legend()
ax[0].grid(True, alpha=0.3)

ax[1].plot(xgb_recalls, xgb_precisions, color='steelblue')
ax[1].scatter(xgb_recalls[xgb_best_idx], xgb_precisions[xgb_best_idx], color='red', s=100, zorder=5, label=f'Best F1 ({xgb_best_threshold:.3f})')
ax[1].set_xlabel('Recall')
ax[1].set_ylabel('Precision')
ax[1].set_title('XGBoost: Precision-Recall Curve')
ax[1].legend()
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Apply optimal threshold
y_pred_xgb_tuned = (y_proba_xgb >= xgb_best_threshold).astype(int)
print("\n--- XGBoost with Tuned Threshold ---")
print(classification_report(y_test, y_pred_xgb_tuned))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb_tuned))
````

### Consumer Behavior Model

```
We want to build a model without using site analytics, which means removing:
    PageValue, 
    ExitRates,
    BounceRates

Hopefully this will isolate the impact of consumer behavior:
    - How many pages they visit (Administrative, Informational, Product)
    - How long they spend on each type of page
    - When they shop (Month, Weekend, SpecialDay)
    - How they arrive (TrafficType, VisitorType)
    - Their technology/location (OS, Browser, Region)

We will train both Random Forest and XGBoost on this reduced feature set, 
compare performance, and extract actionable consumer behavior insights.

    
We  can expect lower accuracy than our full-feature models since we're removing the 
strongest predictor (PageValue).

However, the remaining features provide more insight into what ecommerce sites can actually change to improve revenue/conversion rate. 
```

````python
# Define site-level metrics to EXCLUDE
# These are website analytics metrics, not individual consumer behavior
site_metric_keywords = ['PageValue', 'ExitRate', 'BounceRate']

# Get behavior-only columns from X_train (already one-hot encoded)
behavior_cols = [col for col in X_train.columns 
                 if not any(keyword.lower() in col.lower() for keyword in site_metric_keywords)]

print(f"Total features (full model): {X_train.shape[1]}")
print(f"Behavior-only features: {len(behavior_cols)}")
print(f"\nRemoved features:")
removed = [col for col in X_train.columns if col not in behavior_cols]
for r in removed:
    print(f"  - {r}")

# Create behavior-only train/test sets
X_train_behavior = X_train[behavior_cols]
X_test_behavior = X_test[behavior_cols]

print(f"\nBehavior-only train shape: {X_train_behavior.shape}")
print(f"Behavior-only test shape: {X_test_behavior.shape}")
````

#### Random Forest on Consumer Behavior Features

````python
# Random Forest on Behavior-Only Features
# Using RandomizedSearchCV with the same tight parameter ranges from our full-feature tuning

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve, f1_score

rf_behavior = RandomForestClassifier(random_state=42)

rf_behavior_param_grid = {
    'n_estimators': [300, 400, 500],
    'max_depth': [18, 20, 25],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_features': ['sqrt', 0.3, 0.5],
    'min_samples_split': [2, 5, 10],
    'class_weight': ['balanced'],
}

rf_behavior_search = RandomizedSearchCV(
    estimator=rf_behavior,
    param_distributions=rf_behavior_param_grid,
    cv=5, n_jobs=-1, scoring='f1',
    n_iter=50, random_state=42
)
rf_behavior_search.fit(X_train_behavior, y_train)

print(f"Best Hyperparameters: {rf_behavior_search.best_params_}")
print(f"Best CV F1 Score: {rf_behavior_search.best_score_:.4f}\n")

# Evaluate on test set
y_pred_rf_behavior = rf_behavior_search.predict(X_test_behavior)
print("Random Forest (Behavior-Only) Classification Report:")
print(classification_report(y_test, y_pred_rf_behavior))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf_behavior))

# Threshold tuning
y_proba_rf_behavior = rf_behavior_search.predict_proba(X_test_behavior)[:, 1]
rf_b_precisions, rf_b_recalls, rf_b_thresholds = precision_recall_curve(y_test, y_proba_rf_behavior)
rf_b_f1_scores = 2 * (rf_b_precisions[:-1] * rf_b_recalls[:-1]) / (rf_b_precisions[:-1] + rf_b_recalls[:-1] + 1e-10)
rf_b_best_idx = np.argmax(rf_b_f1_scores)
rf_b_best_threshold = rf_b_thresholds[rf_b_best_idx]

print(f"\nOptimal Threshold: {rf_b_best_threshold:.3f}")
y_pred_rf_behavior_tuned = (y_proba_rf_behavior >= rf_b_best_threshold).astype(int)
print("\n--- RF Behavior-Only with Tuned Threshold ---")
print(classification_report(y_test, y_pred_rf_behavior_tuned))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf_behavior_tuned))
````

```
As we can see here the accuracy suffers as expected, with accuracy being about 10-20% lower. However, with the tuned threshold we are getting a recall of 63% and correctly identifying 259 instances of Revenue.

If we compare this to our Optimized RF model(without Threshold Tuning) we only captured 239 instances of Revenue. This appears promising, as we are able to predict more revenue generating session correctly without using Site Metric features(including the powerful PageValue Features) as the parametrized RF model from earlier.

XGBoost on Consumer Behavior Features

````python
# XGBoost on Behavior-Only Features
from xgboost import XGBClassifier

scale_pos_behavior = (y_train == 0).sum() / (y_train == 1).sum()

xgb_behavior = XGBClassifier(eval_metric='logloss', random_state=42)

xgb_behavior_param_grid = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9, 1.0],
    'colsample_bytree': [0.5, 0.7, 0.8, 1.0],
    'min_child_weight': [1, 3, 5, 7],
    'gamma': [0, 0.1, 0.3, 0.5],
    'scale_pos_weight': [1, scale_pos_behavior],
}

xgb_behavior_search = RandomizedSearchCV(
    estimator=xgb_behavior,
    param_distributions=xgb_behavior_param_grid,
    cv=5, n_jobs=-1, scoring='f1',
    n_iter=100, random_state=42, verbose=1
)
xgb_behavior_search.fit(X_train_behavior, y_train)

print(f"\nBest Hyperparameters: {xgb_behavior_search.best_params_}")
print(f"Best CV F1 Score: {xgb_behavior_search.best_score_:.4f}\n")

# Evaluate on test set
y_pred_xgb_behavior = xgb_behavior_search.predict(X_test_behavior)
print("XGBoost (Behavior-Only) Classification Report:")
print(classification_report(y_test, y_pred_xgb_behavior))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb_behavior))

# Threshold tuning
y_proba_xgb_behavior = xgb_behavior_search.predict_proba(X_test_behavior)[:, 1]
xgb_b_precisions, xgb_b_recalls, xgb_b_thresholds = precision_recall_curve(y_test, y_proba_xgb_behavior)
xgb_b_f1_scores = 2 * (xgb_b_precisions[:-1] * xgb_b_recalls[:-1]) / (xgb_b_precisions[:-1] + xgb_b_recalls[:-1] + 1e-10)
xgb_b_best_idx = np.argmax(xgb_b_f1_scores)
xgb_b_best_threshold = xgb_b_thresholds[xgb_b_best_idx]

print(f"\nOptimal Threshold: {xgb_b_best_threshold:.3f}")
y_pred_xgb_behavior_tuned = (y_proba_xgb_behavior >= xgb_b_best_threshold).astype(int)
print("\n--- XGBoost Behavior-Only with Tuned Threshold ---")
print(classification_report(y_test, y_pred_xgb_behavior_tuned))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb_behavior_tuned))
````

## Research Question 3
### Are there distinct types of browsing sessions, and do they differ in purchase likelihood?
### Can we successfully identify and group distinct patterns of consumer behavior using clustering?

```
To answer this question I will use K-Means clustering on the full dataset using scaled features. 

Firstly, I need to merge the train and test datasets and add their labels back.
Next I need to determine how many clusters I want to create.
    To accomplish this goal I will use both the Elbow and Silhouette Scores
Then I will use the K-Means clustering algorithm over multiple iterations.
When Clusters are sufficiently separated:
    Profile each cluster by examining average features and conversion rates.
    Utilize PCA for dimensionality reductions, hopefully determining what features are really driving the clustering.
```

```
Re-merging the scaled dataset.

````python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

X_all_scaled = pd.concat([X_train_scaled, X_test_scaled], axis=0)
y_all = pd.concat([y_train, y_test], axis=0)

print(f"Combined dataset shape: {X_all_scaled.shape}")
print(f"Revenue distribution:\n{y_all.value_counts()}")
print(f"Revenue rate: {y_all.mean():.3f}")
````

#### Finding optimal K - number of clusters.

````python
# Elbow Method and Silhouette Analysis to determine optimal k
k_range = range(2, 11)
inertias = []
silhouette_scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_all_scaled)
    inertias.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_all_scaled, cluster_labels))
    print(f"k={k}: Inertia={kmeans.inertia_:.0f}, Silhouette={silhouette_scores[-1]:.4f}")

# Plot Elbow and Silhouette side by side
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

ax[0].plot(k_range, inertias, marker='o', color='steelblue')
ax[0].set_xlabel('Number of Clusters (k)')
ax[0].set_ylabel('Inertia (Within-Cluster Sum of Squares)')
ax[0].set_title('Elbow Method')
ax[0].grid(True, alpha=0.3)

ax[1].plot(k_range, silhouette_scores, marker='o', color='darkorange')
ax[1].set_xlabel('Number of Clusters (k)')
ax[1].set_ylabel('Silhouette Score')
ax[1].set_title('Silhouette Analysis')
ax[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

best_k = list(k_range)[np.argmax(silhouette_scores)]
print(f"\nBest k by Silhouette Score: {best_k} (score={max(silhouette_scores):.4f})")
````

```
The elbow method indicates a plateu around 7 clusters.

However, the highest silhouette score is achieved at k=5, and declines after that.

Therefore we should use 5 clusters for our K-Means model.
``

````python
# Fit final K-Means with the best k from silhouette analysis
kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
cluster_labels = kmeans_final.fit_predict(X_all_scaled)

print(f"Final K-Means with k={best_k}")
print(f"Inertia: {kmeans_final.inertia_:.0f}")
print(f"Silhouette Score: {silhouette_score(X_all_scaled, cluster_labels):.4f}")
print(f"\nCluster sizes:")
cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
for cluster, count in cluster_counts.items():
    print(f"  Cluster {cluster}: {count} sessions ({count/len(cluster_labels)*100:.1f}%)")
````

### Limitations

Several limitations should be considered when interpreting these findings:

1. **Single Dataset, Unknown Origin:** This analysis is based on one dataset of 12,330 sessions from an unspecified e-commerce site. We do not know the geographic scope, product category, site design, or time period beyond what the features encode. Results may not generalize to other e-commerce platforms, industries, or regions.

2. **PageValues as a Potential Data Leakage Concern:** PageValues is computed from the revenue generated by each page — making it partially derived from the target variable. While it is technically a session-level feature computed from page-level analytics, its strong correlation with Revenue (~70% of XGBoost importance) suggests it may encode information about the outcome rather than purely predictive input. Our behavior-only models address this concern, but the full-feature model results should be interpreted with this caveat.

3. **Class Imbalance:** The ~85/15 split between non-revenue and revenue sessions means that models can achieve ~85% accuracy by predicting all sessions as non-revenue. We mitigated this with class_weight='balanced' (RF) and scale_pos_weight (XGBoost), threshold tuning, and focusing on F1/recall rather than accuracy alone. However, the imbalance likely still affects model calibration and generalizability.

4. **Feature Granularity:** The dataset provides aggregate session-level counts and durations but not the sequence of page visits. We cannot analyze clickstream patterns, time between pages, or the order in which users engage with different page types. Richer sequential data might reveal additional behavioral patterns.

5. **Temporal Assumptions:** We treat all sessions as independent observations, but browsing behavior likely varies by time of day, day of week, and broader economic conditions. The dataset contains month-level temporal information but no timestamps, limiting our ability to model finer temporal patterns.

6. **Clustering Assumptions:** K-Means assumes spherical, equally-sized clusters and uses Euclidean distance on standardized features.
