````python
# Conduct Hypothesis Test on Weekend proportion to determine if likelihood of session occuring on weekend is significantly different from 2/7 (28.57%) at 95% confidence level
from statsmodels.stats.proportion import proportions_ztest
# Number of sessions that occurred on the weekend
weekend_sessions = ecom_df['Weekend'].sum()
# Total number of sessions
total_sessions = len(ecom_df)
# Proportion of weekend sessions
weekend_proportion = weekend_sessions / total_sessions
# Perform one-sample z-test against the null hypothesis that the true proportion is 2/7
z_stat, p_value = proportions_ztest(count=weekend_sessions, nobs=total_sessions, value=2/7, alternative='two-sided')
print(f"Weekend Proportion: {weekend_proportion:.4f}")
print(f"Z-statistic: {z_stat:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: The proportion of weekend sessions is significantly different from 2/7.")
else:
    print("Fail to reject the null hypothesis: The proportion of weekend sessions is not significantly different from 2/7.")
````

## Outlier

### Outlier Detection

```

Firstly, I want to calculate the IQR's of each column and identify outliers outside of these bounds:
    [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR]

#### Outliers for Numerical Variables

````python
# Identifying outliers with IQR
iqr_outlier_dict = {}

for col in num_cols:

    Q1 = ecom_df[col].quantile(0.25)
    Q3 = ecom_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = ecom_df[(ecom_df[col] < lower_bound) | (ecom_df[col] > upper_bound)]

    iqr_outlier_dict[col] = len(outliers)
````

````python
# Identifying outliers with Z-Scores
z_scores_outlier_dict = {}

for col in num_cols:
    
    z_scores = np.abs(stats.zscore(ecom_df[col]))
    outliers = ecom_df[z_scores > 3]

    z_scores_outlier_dict[col] = len(outliers) 
````

````python
#Comparing No. Outliers identified by IQR vs Z-Score for each numeric column
print("\nComparing No. Outliers identified by IQR vs Z-Score (IQR - Z-Score):")
for col in num_cols:
    iqr_outliers = iqr_outlier_dict[col]
    zscore_outliers = z_scores_outlier_dict[col]
    print(f"\n{col}: Difference = {iqr_outliers - zscore_outliers}")
````

```
It appears Z-score is a much more conservative estimate of outliers.
I think it would be better to work with Z-score outliers to retain as much original data as possible.

```

````python
print("=== Z-Score Outliers (|z| > 3) ===")
for col in num_cols:
    z = np.abs(stats.zscore(ecom_df[col]))
    mask = z > 3
    n_out = mask.sum()
    ratio = n_out / len(ecom_df)
    if n_out > 0:
        print(f"\n{col}: {n_out} outliers ({ratio:.4f})")
        print(f"  Mean={ecom_df[col].mean():.2f}, Std={ecom_df[col].std():.2f}")
        print(f"  Outlier mean={ecom_df[col][mask].mean():.2f}, max={ecom_df[col][mask].max():.2f}")
    else:
        print(f"\n{col}: 0 outliers")

print(f"\nTotal rows: {len(ecom_df)}")

# Also check overlap: how many rows are outliers in ANY column
z_score_outlier_mask_all = pd.DataFrame()
for col in num_cols:
    z_score_outlier_mask_all[col] = np.abs(stats.zscore(ecom_df[col])) > 3
any_outlier = z_score_outlier_mask_all.any(axis=1)
print(f"\nRows that are outliers in at least one column: {any_outlier.sum()} ({any_outlier.mean():.4f})")

# Revenue breakdown of outlier rows
print(f"\nRevenue breakdown of outlier rows:")
print(ecom_df.loc[any_outlier, 'Revenue'].value_counts())
print(f"\nRevenue breakdown of non-outlier rows:")
print(ecom_df.loc[~any_outlier, 'Revenue'].value_counts())
````
