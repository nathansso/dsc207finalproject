````python
#Counts of Weekend
ecom_df['Weekend'].value_counts()

#Proportion of Weekend as a fraction over 7
weekend_fraction = ecom_df['Weekend'].value_counts(normalize=True)*7

print("Proportion of Weekend (as a fraction over 7):")
print(f" Weekend: {weekend_fraction[True]} / 7, Weekday: {weekend_fraction[False]} / 7")
````

````python
# Conduct Hypothesis Test on Weekend proportion to determine if likelihood of session occuring on weekend is significantly different from 2/7 (28.57%) at 95% confidence level
from statsmodels.stats.proportion import proportions_ztest
# Number of sessions that occurred on the weekend
weekend_sessions = ecom_df['Weekend'].sum()
# Total number of sessions
total_sessions = len(ecom_df)
# Proportion of weekend sessions
weekend_proportion = weekend_sessions / total_sessions
# Perform one-sample z-test against the null hypothesis that the true proportion is 2/7
z_stat, p_value = proportions_ztest(count=weekend_sessions, nobs=total_sessions, value=2/7, alternative='two-sided')
print(f"Weekend Proportion: {weekend_proportion:.4f}")
print(f"Z-statistic: {z_stat:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: The proportion of weekend sessions is significantly different from 2/7.")
else:
    print("Fail to reject the null hypothesis: The proportion of weekend sessions is not significantly different from 2/7.")
````

````python
month_order = [
    "Feb", "Mar", "May", "June",
    "Jul", "Aug", "Sep", "Oct",
    "Nov", "Dec"
]
````

````python
## Examine revenue vs non-revenue distribution for categorical columns
## normalized within each category (conversion rate)

sns.set_style("whitegrid")

fig, axes = plt.subplots(2, 4, figsize=(30, 15))
axes = axes.flatten()

cat_no_revenue = [col for col in cat_cols if col != "Revenue"]

for i, col in enumerate(cat_no_revenue):
    ax = axes[i]

    # Normalize within each category
    prop_table = (
        pd.crosstab(ecom_df[col], ecom_df["Revenue"], normalize="index")
        .rename(columns={0: "No", 1: "Yes", False: "No", True: "Yes"})
        .reset_index()
        .melt(id_vars=col, var_name="Revenue", value_name="Proportion")
    )

    sns.barplot(
        data=prop_table,
        x=col,
        y="Proportion",
        hue="Revenue",
        palette={"No": "red", "Yes": "green"},
        order=month_order if col == "Month" else None,
        ax=ax
    )

    ax.set_title(f"Conversion rate by {col}", fontsize=12)
    ax.set_ylabel("Proportion", fontsize=10)
    ax.set_xlabel(col, fontsize=10)
    ax.set_ylim(0, 1)
    ax.tick_params(axis='x', rotation=30)

plt.tight_layout(pad=3.0, w_pad=2.5, h_pad=2.5)
plt.show()
````
